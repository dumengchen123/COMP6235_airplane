```python
import pandas as pd
import pickledb    #nosql DB
import os
```

### Process the flight data to get the training dataset

```python
weather_data = pd.read_excel('./dataset/dataset_raw_weather.xlsx', 
                             usecols=['date', 'Maximum', 'Minimum', 'Average', 'Departure', 'Precipitation'])
delay_data = pd.read_csv('./dataset/dataset_raw_delay.csv', usecols=['date', 'delay_mins'])

weather_data['date'] = pd.to_datetime(weather_data['date']).dt.strftime('%Y-%m-%d')
delay_data['date'] = pd.to_datetime(delay_data['date'], format='%m/%d/%Y').dt.strftime('%Y-%m-%d')
delay_data['id'] = range(1, len(delay_data) + 1)
merged_data = pd.merge(delay_data,weather_data , on='date', how='left')
```

```python
for column in merged_data.columns:
    if column != 'date':
        merged_data[column] = pd.to_numeric(merged_data[column], errors='coerce')
    
merged_data.fillna(0, inplace=True)
```

```python
merged_data
```

```python
if os.path.exists('merged_dataset.db'):
    os.remove('merged_dataset.db')
if os.path.exists('datasets.db'):
    os.remove('datasets.db')

    
db = pickledb.load('merged_dataset.db', False)
for index, row in merged_data.iterrows():
    db.set(str(row['id']), row.to_dict()) 
db.dump()
db = pickledb.load('merged_dataset.db', False)

data = []
for key in db.getall():
    data.append(db.get(key))
df = pd.DataFrame(data)
```

```python
df
```

```python
df.drop('id', axis=1, inplace=True)
filtered_df = df[df['delay_mins'] > 0]
agg_df = filtered_df.groupby('date').agg({
    'delay_mins': 'count',
    'Maximum': 'first',  
    'Minimum': 'first', 
    'Average': 'first',  
    'Departure': 'first', 
    'Precipitation': 'first' 
}).reset_index()
agg_df.rename(columns={'delay_mins': 'delay_times'}, inplace=True)

agg_df['label'] = (agg_df['delay_times'] > 50).astype(int)

agg_df.drop('delay_times', axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(agg_df, test_size=0.2, random_state=42)


db = pickledb.load('datasets.db', False)

db.set('train_dataset', train_df.to_dict(orient='records'))

db.set('test_dataset', test_df.to_dict(orient='records'))

db.dump()
```

```python
train_df.head()
```

### Date-Preprocess

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
train_df[['Maximum', 'Minimum', 'Average', 'Departure', 'Precipitation']] = scaler.fit_transform(train_df[['Maximum', 'Minimum', 'Average', 'Departure', 'Precipitation']])
test_df[['Maximum', 'Minimum', 'Average', 'Departure', 'Precipitation']] = scaler.transform(test_df[['Maximum', 'Minimum', 'Average', 'Departure', 'Precipitation']])
```

### Modeling

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
from sklearn.metrics import accuracy_score, f1_score
```

```python
X_train = train_df[['Maximum', 'Minimum', 'Average', 'Departure', 'Precipitation']]
y_train = train_df['label']
X_test = test_df[['Maximum', 'Minimum', 'Average', 'Departure', 'Precipitation']]
y_test = test_df['label']

# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

# Support Vector Machine
svm_model = SVC()
svm_model.fit(X_train, y_train)

# Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

# LightGBM
lgb_model = lgb.LGBMClassifier()
lgb_model.fit(X_train, y_train)

lr_pred = lr_model.predict(X_test)
svm_pred = svm_model.predict(X_test)
rf_pred = rf_model.predict(X_test)
lgb_pred = lgb_model.predict(X_test)

lr_accuracy = accuracy_score(y_test, lr_pred)
svm_accuracy = accuracy_score(y_test, svm_pred)
rf_accuracy = accuracy_score(y_test, rf_pred)
lgb_accuracy = accuracy_score(y_test, lgb_pred)

lr_f1 = f1_score(y_test, lr_pred)
svm_f1 = f1_score(y_test, svm_pred)
rf_f1 = f1_score(y_test, rf_pred)
lgb_f1 = f1_score(y_test, lgb_pred)
```

```python
print("Logistic Regression - Accuracy:", lr_accuracy, "F1:", lr_f1)
print("Support Vector Machine - Accuracy:", svm_accuracy, "F1 :", svm_f1)
print("Random Forest - Accuracy:", rf_accuracy, "F1 :", rf_f1)
print("LightGBM - Accuracy:", lgb_accuracy, "F1 :", lgb_f1)
```

### Feature Importance Analysis

```python
import matplotlib.pyplot as plt

lgb.plot_importance(lgb_model, max_num_features=10, figsize=(10, 6), importance_type='split')
plt.title("Feature Importance Ranking")
plt.show()

feature_importance = pd.DataFrame({'Feature': X_train.columns, 'Importance': lgb_model.feature_importances_})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
```
